```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
# Read scraped data file into Pandas
data_df = pd.read_csv('akdor_oil_production.csv')
```

```python
data_df = data_df.rename(columns={'1':'PrudhoeBay','2':'Kuparuk','3':'Endicott','4':'Lisburne','5':'Alpine', '6':'ANS','7':'Inventories'})
```

```python
data_df.dropna(how='any')
```

```python
data_df = data_df[data_df["Date"].str.contains("Date")==False]
data_df.dtypes
```

```python
data_df['Date']= pd.to_datetime(data_df['Date'])
data_df['PrudhoeBay'] = pd.to_numeric(data_df['PrudhoeBay'],errors='coerce')
data_df['Kuparuk'] = pd.to_numeric(data_df['Kuparuk'],errors='coerce')
data_df['Endicott'] = pd.to_numeric(data_df['Endicott'],errors='coerce')
data_df['Lisburne'] = pd.to_numeric(data_df['Lisburne'],errors='coerce')
data_df['Alpine'] = pd.to_numeric(data_df['Alpine'],errors='coerce')
data_df['ANS'] = pd.to_numeric(data_df['ANS'],errors='coerce')
data_df['Inventories'] = pd.to_numeric(data_df['Inventories'],errors='coerce')
data_df.dtypes
```

```python
sns.pairplot(data_df)
```

```python
sns.relplot(x=data_df['Date'], y=data_df['PrudhoeBay'], kind='line')
```

```python
# help found at https://towardsdatascience.com/working-with-datetime-in-pandas-dataframe-663f7af6c587
data_df['Year'] = data_df['Date'].dt.year
grouped_year = data_df.groupby('Year').sum()
grouped_year
```

```python
sns.displot(grouped_year['PrudhoeBay'], x='Year', kind='hist')
```
